```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv

%matplotlib notebook
```

```python
# Power law function

x = np.linspace(10, 5000, 100)

# Power law function values
y = 1e7 * ((1 / x)**0.62)

plt.figure(figsize=(10, 5))

plt.title('Twitter user follower amount based on their rank')

plt.xlabel('User rank')
plt.ylabel('Amount of followers')

plt.plot(x,y, 'r')

plt.savefig('../images/power_law.png')

plt.show()
```

```python
# Network simulation spread (Work In Progress!)

dff = pd.read_csv(f'../data/trump_chanting_verified.csv')
dff = dff[:400]

yData = np.cumsum(dff['total_tweets'])

df = pd.read_csv('../server/data/trump_simulation_1000.csv', header=None)

x_common = np.linspace(0,400,400)

plt.figure(figsize=(10, 5))

plt.title('Spread of 1000 network simulations (trump_chanting_verified)')

# y = np.stack(I_simulations)

ymax = np.max(df, axis=0)
ymin = np.min(df, axis=0)

plt.plot(yData, 'tab:orange', label='Original Data')

plt.plot(x_common, ymax, 'tab:blue', label='1000 Network Simulations')
plt.plot(x_common, ymin, 'tab:blue')

plt.fill_between(x_common, ymax, ymin, facecolor=(0, 0, 1, 0.2), hatch ='/', edgecolor='tab:blue', linewidth=2)
plt.xlabel('Time step')
plt.ylabel('Number of tweets')
plt.legend()

plt.savefig('../images/trump_1000_network_simulations.png')

plt.show()
```
